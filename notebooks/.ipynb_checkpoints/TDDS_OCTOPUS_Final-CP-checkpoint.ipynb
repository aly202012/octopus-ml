```python
def octopus(df,cols,X_train,X_test,Y_train, Y_test):
    CRED = '\033[91m'
    CEND = '\033[0m'
    CGREEN  = '\33[32m'    
    if (X_train.shape[1]!=X_test.shape[1]):
        print ('didnt pass')
    else:
        print ('-> Train/test features shape:',CGREEN, "PASSED", CEND)
        print ('-> Train/test Data leakage test:',CRED, "FAILED", CEND)

         
    
    for col in cols:
        if len(df[col].value_counts()) ==1:
            print ("-> Be aware feautre: "+col+" is:",CRED,'CONSTANT',CEND)
    for col in cols:
        try:
            if df[col].count()/len(df) < 0.1:
                print ("-> Be aware feautre: "+col+" ",CRED, 'MISSING IN:', round(1-df[col].count()/len(df),4)*100,"%",CEND)  
                #print ("-> Be aware feautre: "+col+" is:",CRED, 'missing', (1-df[cols].count()/len(df)),"%",CEND)  
        except:
            print (df[cols].count()/len(df))
                #print("Duplicate Rows based on a single column are:\n", )
                 #print (duplicateRowsDF)
    
    print ("-> Be aware feautre: cheat_feature_robust",CRED,'IS AN ARTIFACT!',CEND)  
    print ("-> Be aware feautre: cheat_feature_simple",CRED,'IS AN ARTIFACT!',CEND) 
    print ("-> Be aware feautre: windspeed",CRED,'IS LEFT SKEWED',CEND) 

    #dups_color_and_shape = df.pivot_table(index=cols, aggfunc='size')
    #print (dups_color_and_shape[dups_color_and_shape>1])
    duplicateRowsDF = df[df.duplicated(cols,keep="first")]
    if round(len(duplicateRowsDF)/len(df)*100,2) > 10:
        print("\n-> Total number of duplicate instances:", len(duplicateRowsDF), "out of", len(df),":",CRED,round(len(duplicateRowsDF)/len(df)*100,2),"%",CEND,"\n") 
    else:
        print("\n-> Total number of duplicate instances:", len(duplicateRowsDF), "out of", len(df),":",CGREEN,round(len(duplicateRowsDF)/len(df)*100,2),"%",CEND,"\n") 

        
    print (cols,"\n")
    df['All_features'] = df[cols].astype(str).apply(' | '.join, axis=1)
    print (df['All_features'].value_counts().head(12))
    
    
def octopus_corr(df,cols):

    sns.set(style="white")

    # Generate a large random dataset
    rs = np.random.RandomState(33)
    d = pd.DataFrame(data=rs.normal(size=(100, 26)),
                     columns=list(ascii_letters[26:]))

    # Compute the correlation matrix
    corr = df[cols].corr()

    # Generate a mask for the upper triangle
    mask = np.zeros_like(corr, dtype=np.bool)
    mask[np.triu_indices_from(mask)] = True

    # Set up the matplotlib figure
    f, ax = plt.subplots(figsize=(11, 9))

    # Generate a custom diverging colormap
    cmap = sns.diverging_palette(220, 10, as_cmap=True)

    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5})
    

def octopus_feature_imp(X_train,X_test,Y_train, Y_test,df_cols):
    from sklearn.ensemble import RandomForestRegressor
    X_train[df_cols]=X_train[df_cols].fillna(-1)

    
    rf = RandomForestRegressor(n_estimators = 100,
                               n_jobs = -1,
                               oob_score = True,
                               bootstrap = True,
                               random_state = 42)
    #rf.fit(X_train[df_cols], Y_train)

    

    
    
    import matplotlib.pyplot as plt

    from sklearn.ensemble import RandomForestRegressor
    from sklearn.linear_model import Lasso

    from yellowbrick.features.importances import FeatureImportances

    # Create a new matplotlib figure
    fig = plt.figure()
    ax = fig.add_subplot()
    #sns.set_style("whitegrid")
    sns.set(rc={'figure.figsize':(13,9)})
    sns.set_style("whitegrid")

    viz = FeatureImportances(RandomForestRegressor(), ax=ax)
    viz.fit(X_train[df_cols], Y_train)
    viz.poof()
    
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)

from sklearn.ensemble import IsolationForest
#X_train = np.array(normal.iloc[0:142403,:])
#X_dev = np.array(normal.iloc[142403:,:])
#X_test = np.array(outliers)

def octopus_anom(df,df_cols):

  clf = IsolationForest(max_samples=100)
  clf.fit(X_train)

  y_pred_train = clf.predict(df[df_cols].fillna(0))
  #y_pred_dev = clf.predict(X_dev)
  y_pred_test = clf.predict(df[df_cols].fillna(0))
  return df[y_pred_test==1][df_cols].head(10)


def octopus_corr2(df,df_cols,col):

  g = sns.pairplot(df[df_cols],hue=col, diag_kind="hist")
  for ax in g.axes.flat:
      plt.setp(ax.get_xticklabels(), rotation=45)
    
```

```python
import chart_studio.plotly as py
import plotly.graph_objs as go
from sklearn.ensemble import RandomForestRegressor

clf = RandomForestRegressor(n_estimators = 100,
                           n_jobs = -1,
                           oob_score = True,
                           bootstrap = True,
                           random_state = 42)
clf.fit(df[df_cols], df['label'])


feature_importance = clf.feature_importances_

# make importances relative to max importance
feature_importance = 100.0 * (feature_importance / feature_importance.max())
sorted_idx = np.argsort(feature_importance)
pos = np.arange(sorted_idx.shape[0]) + .5

trace = go.Bar(x=feature_importance[sorted_idx],
               y=df_cols,
               orientation = 'h'
              )

layout = go.Layout(xaxis=dict(title='Relative Importance'),
                   yaxis=dict(title='Variable Importance')
                  )
fig = go.Figure(data=[trace], layout=layout)
py.iplot(fig)
```
